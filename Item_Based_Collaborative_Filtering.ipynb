```
import pandas as pd
import requests as rq


r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('http://files.grouplens.org/datasets/movielens/ml-100k/u.data', sep='\t', names = r_cols, usecols=range(3))
ratings = ratings.append({'user_id': 0, 'movie_id': 172, 'rating': 5}, ignore_index=True)
ratings = ratings.append({'user_id': 0, 'movie_id': 133, 'rating': 1}, ignore_index=True)
ratings = ratings.append({'user_id': 0, 'movie_id': 50, 'rating': 5}, ignore_index=True)

m_cols = ['movie_id', 'title']
movies = pd.read_csv('http://files.grouplens.org/datasets/movielens/ml-100k/u.item', sep='|', names = m_cols, usecols=range(2), encoding = "ISO-8859-1")

ratings = pd.merge(movies, ratings)
ratings.head()
```

```
userRatings = ratings.pivot_table(index=['user_id'], columns=['title'], values='rating')
userRatings.head()
```

```
corrMatrix = userRatings.corr()
corrMatrix.head()
```

```
corrMatrix = userRatings.corr(method='pearson', min_periods=100)
corrMatrix.head()
```

```
myRatings = userRatings.loc[0].dropna()
myRatings.head()
```

```
similaritiesCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
  print(f'Appending sims for {myRatings.index[i]}...')
  # Retrieve similar movies to this one that I rated
  sims = corrMatrix[myRatings.index[i]].dropna()
  # Now scale its similarity by how well I rated this movie
  sims = sims.map(lambda x: x * myRatings[i])
  # Add the score to the list of similarity candidates
  similaritiesCandidates = similaritiesCandidates.append(sims)
  
print("sorting...")
similaritiesCandidates.sort_values(inplace=True, ascending=False)
print(similaritiesCandidates.head())
```

```
simCandidates = similaritiesCandidates.groupby(similaritiesCandidates.index).sum()
simCandidates.sort_values(inplace=True, ascending=False)
simCandidates.head(10)
```

```
filteredSims = simCandidates.drop(myRatings.index)
filteredSims.head(10)
```
