```python
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import numpy as np
from pylab import *

R2_test = []
R2_train = []

np.random.seed(2)

pageSpeeds = np.random.normal(3.0, 1.0, 100)
purchaseAmount = np.random.normal(50.0, 30.0, 100) / pageSpeeds


plt.scatter(pageSpeeds, purchaseAmount)
```

```python
trainX = pageSpeeds[:80]
testX = pageSpeeds[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]

scatter(trainX, trainY, color='b')
scatter(testX, testY, color='r')
```

```python
X = np.array(pageSpeeds)
Y = np.array(purchaseAmount)

trainX = np.array(trainX)
trainY = np.array(trainY)

testX = np.array(testX)
testY = np.array(testY)
```

```python
fits = []
r2_test = []
r2_train = []

deg = range(1,10)

for i in deg:
    p = np.poly1d(np.polyfit(trainX, trainY, i))
    fits.append(p)
    r2_train.append(r2_score(y, p(x)))
    r2_test.append(r2_score(testY, p(testX)))
```

```python
plt.plot(deg, r2_train)
plt.plot(deg, r2_test)
axes = plt.axes()
axes.set_xlim([0,10])
axes.set_ylim([0, 1])
plt.show()
```

```python
R2 = np.multiply(r2_train, r2_test)
_, maxIdx = np,argmax(R2)
selectedDeg = deg[maxIdx]
selectedDeg
```
